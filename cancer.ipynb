KNN [K- Nearest Neighbour]
** **
using knn model to predict breast cancer and visualizing it

```python
#imports
import numpy as np 
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

Importing data set

```python
cancer_dataset = pd.read_csv("breast-cancer.csv")
cancer_dataset.head()
```

Feature selection


```python
cancer_dataset['diagnosis']=cancer_dataset['diagnosis'].map({'M':1, 'B':0})
cancer_dataset
```

```python
y = cancer_dataset['diagnosis']

x = cancer_dataset
x = x.drop('diagnosis',axis =1)
print(cancer_dataset.shape)
print(x.shape)
print(y.shape)
```

Train Test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state= 20,stratify=y)
print(x_train.shape,y_train.shape)
```

Model Building [Using KNN model]

```python
knn = KNeighborsClassifier(n_neighbors= 6)
knn.fit(x_train,y_train)
```

Accuracy calculate

```python
y_predict = knn.predict(x_test)
accuracy = accuracy_score(y_test,y_predict)
accuracy
```

Visualizing the outcome

```python
cm = confusion_matrix(y_test,y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm , display_labels=['Beingin(0)','Malignant(1)'])
disp.plot(cmap='Blues')
```
